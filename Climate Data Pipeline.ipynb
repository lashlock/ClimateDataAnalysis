```python
#Want to drop the first row
newDF2009 = df2009.drop(0)

newDF2010 = df2010.drop(0)

newDF2011 = df2011.drop(0)

newDF2012 = df2012.drop(0)

newDF2013 = df2013.drop(0)

newDF2014 = df2014.drop(0)

newDF2015 = df2015.drop(0)

newDF2016 = df2016.drop(0)

newDF2017 = df2017.drop(0)

newDF2018 = df2018.drop(0)

newDF2019 = df2019.drop(0)

newDF2019
```

```python
#Now I want to merge these dataframes together... this should help reduce downstream bulkiness
frames = [newDF2009, newDF2010, newDF2011, newDF2012, newDF2013, newDF2014, newDF2015, newDF2016, newDF2017, newDF2018, newDF2019]
mergeDF = pd.concat(frames)
mergeDF
```

```python
#Alright, now I want to get rid of any unnecessary rows
subDF = mergeDF[['#YY','MM','DD','hh','mm','WTMP']]
subDF
subDF.dtypes
```

```python
#Coerce python into thinking water temperature is numeric using dictionary
convert_dict = {'WTMP': float,
               } 
  
subDF = subDF.astype(convert_dict) 
print(subDF.dtypes)
```

```python
#What is the minimum value for water temperature in this data frame
subDF.loc[subDF['WTMP'].idxmin()]
#This gives you min value by year
```

```python
#What is the max value for water temperature in this data frame
subDF.loc[subDF['WTMP'].idxmax()]
#Need to get rid of 999 values - This is ndbc version of n/a or no data
```

```python
#Getting rid of anything that is >= 999 in our dataframe
subDF.drop( subDF[ subDF['WTMP'] >= 999 ].index , inplace=True)
```
